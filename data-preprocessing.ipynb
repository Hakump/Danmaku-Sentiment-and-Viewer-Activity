```python
def get_filepath_by_index(index, datapath="/Users/kesong/Downloads/VideoIC/data/data/comments/"):
  v_class = info[info.index == index]['v_class'].values[0]
  filepath = v_class+'/'+str(index)+'.json'
  return filepath, datapath+filepath
```

```python
_, filepath = get_filepath_by_index(10582565)
```

```python
filepath
```

```python
def get_df_by_filename(filename):
  with open(filename, 'r') as f:
    file0 = json.load(f)
  rows = []
  for i,(k,v) in enumerate(file0.items()):
    for danmuku in v:
      rows.append([k, danmuku])
  df = pd.DataFrame(rows, columns=['time', 'text'])
  return df
```
